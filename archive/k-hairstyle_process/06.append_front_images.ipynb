# Append Front View Images

This script is used to merge the exported front view images to the already-existing multi_view_hq dataset

```python
import os
import json
import tqdm
import shutil
```

```python
fv_image_rootdir = '/data/multi_view_hq_fv'
dst_rootdir = '/data/PanoHeadData/multi_view_hq'
```

## 1. update json

```python
orgn_dataset_json = os.path.join(dst_rootdir, 'dataset.json')
fv_dataset_json = os.path.join(fv_image_rootdir, 'dataset.json')
with open(orgn_dataset_json, 'r') as f:
    orgn_dataset = json.load(f)
print('orgn_dataset: ', len(orgn_dataset))
if len(orgn_dataset) != 438541:
    with open(fv_dataset_json, 'r') as f:
        fv_dataset = json.load(f)
    print('fv_dataset: ', len(fv_dataset))

    # update dataset
    orgn_dataset.update(fv_dataset)
    print('new_dataset: ', len(orgn_dataset)) 

    # save dataset
    with open(orgn_dataset_json, 'w') as f:
        json.dump(orgn_dataset, f, indent=4)
```

## 2. Copy images

```python
target_name = 'align_images'
model_dirs = os.listdir(os.path.join(fv_image_rootdir, target_name))
for model_dir in sorted(model_dirs):
    src_image_dir = os.path.join(fv_image_rootdir, target_name, model_dir)
    dst_image_dir = os.path.join(dst_rootdir, target_name, model_dir)

    os.makedirs(dst_image_dir, exist_ok=True)
    for image_name in tqdm.tqdm(os.listdir(src_image_dir)):
        src_image_path = os.path.join(src_image_dir, image_name)
        dst_image_path = os.path.join(dst_image_dir, image_name)
        shutil.copy(src_image_path, dst_image_path)
```
